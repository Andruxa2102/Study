## Домашнее задание начального уровня 

Обучите модель DBSCAN с параметром *min_samples=8*. Визуализируйте результат.

```python
# -- ВАШ КОД ТУТ --

import pickle
import numpy as np
from sklearn.cluster import DBSCAN

with open('data/clustering.pkl', 'rb') as f:
    data_clustering = pickle.load(f)

X = np.array(data_clustering['X'])
Y = np.array(data_clustering['Y'])

DB = DBSCAN(min_samples=8)
DB.fit(X)

from matplotlib import pyplot as plt
%matplotlib inline
plt.scatter(X[:,0], X[:,1], s=40, marker='o', c=DB.labels_)
plt.show;
```
