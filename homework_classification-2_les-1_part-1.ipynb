### 1.5
В колонке Sex замените значения на 0, если пол "male", и на 1, если "female". В колонке Embarked замените параметры "U", "S", "C", "Q" на 0, 1, 2, 3 соответственно. Отбросьте колонки PassengerId, Name, Ticket, Cabin. Выведите первые 20 строк получившегося набора данных.

```python
### YOUR CODE HERE ###

start_df = start_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1)
```

```python
import numpy as np

start_df['Sex'] = np.where(start_df.Sex == 'male', 0, np.where(start_df.Sex == 'female', 1, 255))
start_df['Embarked'] = np.where(start_df.Embarked == 'U', 0,
                  np.where(start_df.Embarked == 'S', 1,
                          np.where(start_df.Embarked == 'C', 2,
                                  np.where(start_df.Embarked == 'Q', 3, np.nan))))
start_df['Embarked'].fillna(start_df['Embarked'].mean(), inplace = True)
```

```python
start_df.head(20)
```

### 1.6
Приступим к построению классификаторов. **Условимся, что, если функция или объект модели имеют параметр random_state, то устанавливаем его равным 17 в каждом из случаев.**

Разделите данные на тренировочный и тестовый датасеты, установив размер тестового как 0.25 (первая колонка Survived является целевой, поэтому необходимо сначала ее отделить от признаков).

```python
from sklearn.model_selection import train_test_split 
```

```python
### YOUR CODE HERE ###

X = start_df.drop(['Survived'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, start_df.Survived, test_size=.25,  random_state = 17)
```

### 1.7
Обучите на полученных выборках несколько классификаторов, которые импортированы ниже.

**Примечание: в методе ближайших соседей используйте количество соседей, равное 5.**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
### YOUR CODE HERE ###

KN = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
GNB = GaussianNB().fit(X_train, y_train)
DTC = DecisionTreeClassifier(random_state=17).fit(X_train, y_train)
LR = LogisticRegression(random_state=17).fit(X_train, y_train)
```

Очевидно, в этом случае использовать метрику accuracy, которую мы применяли до этого, не совсем правильно: она не даст верную оценку классификатору, потому как выживших явно меньше, чем погибших.

Вспомним про такие метрики как точность и полнота. Здесь они подойдут хорошо, так как у нас имеются положительный и отрицательный классы, и определить корректно один из них может оказаться более важным .

### 1.10
Для каждого классификатора выведите матрицу ошибок и самостоятельно рассчитайте метрику recall (расчеты должны присутствовать). Проверьте расчеты, воспользовавшись встроенной метрикой recall_score.

```python
from sklearn.metrics import confusion_matrix, recall_score
```

```python
### YOUR CODE HERE ###

KN_cnf_matrix = confusion_matrix(y_test, KN.predict(X_test))
GNB_cnf_matrix = confusion_matrix(y_test, GNB.predict(X_test))
DTC_cnf_matrix = confusion_matrix(y_test, DTC.predict(X_test))
LR_cnf_matrix = confusion_matrix(y_test, LR.predict(X_test))
```

```python
KN_cnf_matrix
```

```python
GNB_cnf_matrix
```

```python
DTC_cnf_matrix
```

```python
LR_cnf_matrix
```

```python
KN_recall_calculated = KN_cnf_matrix[0,0] / (KN_cnf_matrix[0,0] + KN_cnf_matrix[1,0])
GNB_recall_calculated = GNB_cnf_matrix[0,0] / (GNB_cnf_matrix[0,0] + GNB_cnf_matrix[1,0])
DTC_recall_calculated = DTC_cnf_matrix[0,0] / (DTC_cnf_matrix[0,0] + DTC_cnf_matrix[1,0])
LR_recall_calculated = LR_cnf_matrix[0,0] / (LR_cnf_matrix[0,0] + LR_cnf_matrix[1,0])
```

```python
KN_recall_builtin = recall_score(y_test, KN.predict(X_test))
GNB_recall_builtin = recall_score(y_test, GNB.predict(X_test))
DTC_recall_builtin = recall_score(y_test, DTC.predict(X_test))
LR_recall_builtin = recall_score(y_test, LR.predict(X_test))
```

```python
print('Функция recall_score:\nKN_neighbours:   {}\nGaussianNB:  {}\nDecisionTree:  {}\nLogisticRegression:  {}'.format(KN_recall_builtin, GNB_recall_builtin, DTC_recall_builtin, LR_recall_builtin))
```

```python
print('Расчет из матрицы ошибок:\nKN_neighbours:   {}\nGaussianNB:  {}\nDecisionTree:  {}\nLogisticRegression:  {}'.format(KN_recall_calculated, GNB_recall_calculated, DTC_recall_calculated, LR_recall_calculated))
```
