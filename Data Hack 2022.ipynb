```python
import pandas as pd
from datetime import datetime
import numpy as np
from random import choice
from string import ascii_letters, digits
import time
```

```python
metadata = dict(num=[1, 16], 
                char=[5], 
                varchar=[5], 
                date=['2021.1.1', '2022.1.1'], 
                timestamp=['2021.1.1 6:10:8', '2022.1.1 6:10:8'],
                col_set=(1,2,3,4,5,678),
                num_mask='123####45345', 
                str_mask = ('sdft####sasdf'))
RECORDS_COUNT = 1000
```

```python
metadata
```

```python
datetime.strptime(metadata['date'][0], '%Y.%m.%d').date()
```

```python
datetime.strptime(metadata['timestamp'][0], '%Y.%m.%d %H:%M:%S')
```

```python
df = pd.DataFrame()
cnt = 0

for i in metadata.keys():
    if i == 'num':
        df[str(cnt)] = np.random.randint(metadata[i][0], metadata[i][1]+1, RECORDS_COUNT)
        
    elif i == 'char':
        df[str(cnt)] = np.array([''.join([choice(ascii_letters) 
                                          for i in range(metadata[i][0])]) 
                                 for j in range(RECORDS_COUNT)])
        
    elif i == 'varchar':
        df[str(cnt)] = np.array([''.join([choice(ascii_letters) 
                                          for i in range(1+np.random.choice(metadata[i][0]))]) 
                                 for j in range(RECORDS_COUNT)])
        
    elif i == 'date':
        start = time.mktime(time.strptime(metadata['date'][0], '%Y.%m.%d'))
        finish = time.mktime(time.strptime(metadata['date'][1], '%Y.%m.%d'))
        rand_arr = np.random.random_sample((RECORDS_COUNT,)) * (finish - start)
        df[str(cnt)] = [time.strftime('%Y.%m.%d', time.localtime(i + start)) for i in rand_arr]
        
    elif i == 'timestamp':
        start = time.mktime(time.strptime(metadata['timestamp'][0], '%Y.%m.%d %H:%M:%S'))
        finish = time.mktime(time.strptime(metadata['timestamp'][1], '%Y.%m.%d %H:%M:%S'))
        rand_arr = np.random.random_sample((RECORDS_COUNT,)) * (finish - start)
        df[str(cnt)] = [time.strftime('%Y.%m.%d %H:%M:%S', time.localtime(i + start)) for i in rand_arr]
        
    elif i == 'col_set':
        df[str(cnt)] = np.random.choice(metadata['col_set'], RECORDS_COUNT)

    elif i == 'num_mask':
        lst = [metadata['num_mask']] * RECORDS_COUNT
        lst = [''.join([i if i != '#' else np.random.choice(list(''.join(choice(digits)))) for i in metadata['num_mask']]) for i1 in lst]
        df[str(cnt)] = lst
        
    elif i == 'str_mask':
        lst = [metadata['str_mask']] * RECORDS_COUNT
        lst = [''.join([i if i != '#' else np.random.choice(list(''.join(choice(ascii_letters)))) for i in metadata['str_mask']]) for i1 in lst]
        df[str(cnt)] = lst
        
    cnt += 1
```
