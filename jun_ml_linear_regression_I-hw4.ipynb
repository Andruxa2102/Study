### Домашняя работа урок 4

Даны точки $x$ и значения в этих точках $y$. Нормализуйте $y$ с помощью *z-score* и постройте график зависимости нормализованных значений от $x$. Для графика используйте `.scatter(x,y_tansformed`

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt

x = np.linspace(1,10,num=10)
y = np.array(
    [1.,  3.,  4.,  2., 10.,  5.,  5.,  2.,  5., 10.],
    dtype=np.float32
)

# -- ВАШ КОД ТУТ --

transformed_y = StandardScaler().fit_transform(y.reshape(-1,1)).reshape(-1)
```

```python
from matplotlib import pyplot as plt
plt.scatter(x,y)
plt.ylabel('переменная x')
plt.xlabel('до z-score преобразования');
```

```python
print(transformed_y)
```

```python
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(15,10))
ax1.scatter(x,y)
ax1.set_xlabel('переменная x')
ax1.set_ylabel('переменная y')
ax1.set_title('до z-score преобразования')
ax1.set_ylim(-3, 11)
ax2.scatter(x,transformed_y)
ax2.set_xlabel('переменная x')
ax2.set_ylabel('переменная transformed_y')
ax2.set_title('после z-score преобразования')
ax2.set_ylim(-3, 11);
```
