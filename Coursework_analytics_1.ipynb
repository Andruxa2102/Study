Учитывать, что выполнение домашнего задания может перетекать из одного месяца в другой (такие дз надо включать в общее число прогрессов для всех месяцев, которые покрывает срок выполнения этих дз)

```python
# Создадим набор данных 'dates_df' для хранения данных по началу и окончанию работы в формате Timestamp

dates_df = main.query('is_homework')[['title','course_id','start_date','finish_date', 'is_homework']]
dates_df['finish_date'] = dates_df['finish_date'].fillna('2019-07-30 00:00:00+00:00')
dates_df['start'] = pd.to_datetime(pd.to_datetime(dates_df['start_date'], errors='coerce').dt.strftime('%Y-%m-%d'))
dates_df['finish'] = pd.to_datetime(pd.to_datetime(dates_df['finish_date'], errors='coerce').dt.strftime('%Y-%m-%d'))

# 'months_list' - список месяцев за весь период аналитики.

months_list = [np.datetime64(i, 'D') for i in temp_df.columns]
```

```python
# столбец 'months' нужен для хранения месяцев, в которых происходило выполнение работы. Заполним его

dates_df['months'] = None

# Не придумал, как сделать векторными операциями - использовал apply и это долго.

def fill_months(pd_obj):
    months = []
    start = pd_obj['start']
    finish = pd_obj['finish']
    
    for i in months_list:
        if i > start:
            if i <= (finish + relativedelta.relativedelta(months=1)):
                months.append(i)
            else:
                break
    return months

dates_df['months'] = dates_df.apply(fill_months, axis=1)
```

```python
# 'temp_df' будет хранить id и title курсов и месяцы('months'), в которые проходила работа по этим курсам

temp_df = dates_df.groupby(['course_id', 'title'])['months'].apply(list).reset_index()

# Сделаем один большой одномерный список 'months_list'

def months_calculate(pd_obj):
    months_list = []
    for i in pd_obj['months']:
        months_list.extend(i)
    return months_list

temp_df['months_list'] = temp_df.apply(months_calculate, axis = 1)
```

Чтобы выявить сезонность, посчитать медианное время выполнения домашней работы по месяцам (12 месяцев, январь-декабрь) для каждого курса. 

```python
# temp_df будет хранить время начала и конца каждой работы в каждом курсе и каждом месяце

temp_df = main[['course_id','title','start_date','finish_date']]
temp_df['finish_date'] = temp_df['finish_date'].fillna('2019-08-02 00:00:00+00:00')
temp_df['start_date'] = pd.to_datetime(temp_df['start_date'], errors='coerce').dt.tz_localize(None)
temp_df['finish_date'] = pd.to_datetime(temp_df['finish_date'], errors='coerce').dt.tz_localize(None)
for i in months_list:
    temp_df[i] = np.nan
```

```python
# Заменим время работы на продолжительность работы в каждом месяце на каждом курсе
# Сначала заполним ячейки, где работа заняла полный месяц

temp_df['calc_start_date'] = temp_df.start_date + pd.offsets.MonthEnd(1)
temp_df['calc_finish_date'] = temp_df.finish_date - pd.offsets.MonthBegin(1)

def fill_months_studytimes_1(pd_obj, column):  
    return np.where(((pd.Timestamp(i) > pd_obj['calc_start_date']) & (pd.Timestamp(i) < pd_obj['calc_finish_date'])),
                   i - (pd.Timestamp(i) - pd.offsets.MonthBegin(1)) + timedelta(days=1), pd.Timedelta(0))

for i in months_list[4:]:
    temp_df[i] = fill_months_studytimes_1(temp_df, i)

temp_df = temp_df.drop(['calc_start_date', 'calc_finish_date'], axis=1)


# Потом заполним ячейки, где работа заняла неполные месяца:
def fill_months_studytimes_2(pd_obj):
    for i in months_list: 
        
        # Если работа начата и закончена в пределах одного месяца
        pd_obj[i] = np.where((pd_obj.start_date < i) & \
                              (pd_obj.finish_date < pd_obj.start_date + pd.offsets.MonthEnd(1)) & \
                              (pd_obj.finish_date > pd.Timestamp(i) - relativedelta.relativedelta(months=1)), \
                              pd_obj.finish_date - pd_obj.start_date, pd_obj[i])
        
        # Если работа вышла за пределы одного месяца, то заполним последний неполный месяц
        pd_obj[i] = np.where((pd_obj.finish_date < i) & \
                             (pd_obj.start_date < pd_obj.finish_date - pd.offsets.MonthEnd(1)) & \
                             (pd_obj.finish_date > pd.Timestamp(i) - relativedelta.relativedelta(months=1)), \
                             pd_obj.finish_date - (pd.Timestamp(i) + pd.offsets.MonthBegin(-1)), pd_obj[i])

        # И первый неполный месяц
        pd_obj[i] = np.where((pd_obj.start_date < i) & \
                             (pd_obj.start_date < pd_obj.finish_date - pd.offsets.MonthEnd(1)) & \
                             (pd_obj.start_date > pd.Timestamp(i) - relativedelta.relativedelta(months=1)), \
                              pd.Timestamp(i) - pd_obj.start_date, pd_obj[i])        

fill_months_studytimes_2(temp_df)
```

```python
# months_dict - словарь для соотнесения месяцев-годов из temp_df с номерами месяцев в календаре
# temp2_df - DataFrame для хранения времени работы в каждом месяце на каждом курсе

months_dict = {i: i.astype('object').month for i in months_list}
temp2_df = pd.DataFrame(columns=['course_id', 'title', *map(str, range(1,13))])

for key, value in months_dict.items():
    indexes = temp_df[temp_df[key] > timedelta(0)].index
    temp2_df = temp2_df.append(pd.DataFrame({'course_id': temp_df.iloc[indexes]['course_id'], \
                                             'title': temp_df.iloc[indexes]['title'], \
                                             str(value): temp_df.iloc[indexes][key]}), ignore_index=True)

# Почему-то в одном из столбцов пустые значения приняли значение pd.NaT - это мешает аггрегированию, исправляем
temp2_df['3'] = temp2_df['3'].replace({pd.NaT: np.nan})

# Для расчета медианной переведем все время работы в секунды.
for i in map(str, range(1,13)):
    temp2_df[i][temp2_df[i].notna()] = temp2_df[i][temp2_df[i].notna()].map(lambda x: x.total_seconds())
```

```python
# Получим искомую таблицу с месяцами, курсами и медианным временем работы в каждом курсе за каждый месяц в сутках.

temp3_df = temp2_df.groupby(['course_id','title']).agg(lambda x: x.median()) / 86400
temp3_df
```
