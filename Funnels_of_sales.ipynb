```python
import pandas as pd
import numpy as np
```

```python
main = pd.read_csv('./data/click_stream3.csv', names = ['id', 'page', 'date', 'device', 'sex'], \
                   dtype = {'id': int, 'page': str, 'sex': str}, parse_dates = ['date'], infer_datetime_format = True)
```

```python
main.head()
```

```python
# В разрезе пола и устройства
grouped_all_df = main.groupby([pd.Grouper(key='date', freq='m'), 'page', 'device', 'sex']).id.count()
grouped_all_df.head(16)
```

```python
pd.options.display.max_rows = 64
```

```python
result_df = grouped_all_df.reset_index()
funnels_result = []

for i in grouped_all_df.reset_index().date.unique():
   
    funnels = np.where(grouped_all_df.xs(i).reset_index().device == 'Desktop', \
                       np.where(grouped_all_df.xs(i).reset_index().sex == 'Female', \
                               grouped_all_df.xs(i).values[12] / grouped_all_df.xs(i).values,
                               grouped_all_df.xs(i).values[13] / grouped_all_df.xs(i).values),
                       np.where(grouped_all_df.xs(i).reset_index().sex == 'Female', \
                               grouped_all_df.xs(i).values[14] / grouped_all_df.xs(i).values,
                               grouped_all_df.xs(i).values[15] / grouped_all_df.xs(i).values))
                       
    funnels_result.extend(funnels)

result_df['funnels'] = funnels_result
result_df.head(64)
```

Выводы:

1. В январе конверсия с женских аккаунтов была со всех страниц выше, чем у мужчин.
2. В апреле конверсия женских аккаунтов со всех страниц стала ниже, чем у мужчин.
3. В феврале и марте конверсия с мобильных устройств и женских аккаунтов выше, чем у мужчин на мобилках, зато конверсия на десктопах у мужчин выше, чем у женщин. Исключение - февраль 3_payment_page.
4. Конверсия с мобильных устройств всегда выше, чем с настольных.
